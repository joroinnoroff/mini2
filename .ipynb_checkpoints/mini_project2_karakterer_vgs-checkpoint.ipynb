# Mini-project 2 – Norwegian High School Grade Statistics

**Problem:** There is a public debate about whether standpunkt (teacher) grades are systematically
higher than written exam grades, and whether oral exam grades are often higher than standpunkt.
We explore this gap using data from Udir for Oslo-area schools.

**Data source:** https://www.udir.no/tall-og-forskning/statistikk/statistikk-videregaendeskole/karakterer-vgs/
Export the table as CSV and save it as `udir_karakterer.csv` in the same folder as this notebook.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV exported from Udir statistikkbank
path = "Karakterer_i_videregaaende_skole.csv"
df = pd.read_csv(path)

df.head()
```

```python
# Check how many rows we have and the column data types
print(f"Number of rows: {len(df)}")
print(f"\nColumn types:\n{df.dtypes}")
```

```python
# Check the unique values in the assessment type column
print("Assessment types:", df["Vurderingsform"].unique())

# Check available school years
print("School years:", df["Skoleaar"].unique())
```

```python
# Make sure the grade column is numeric (some cells may be '-' or empty in the Udir export)
df["Snittkarakter"] = pd.to_numeric(df["Snittkarakter"], errors="coerce")

# Drop rows where grade is missing
df = df.dropna(subset=["Snittkarakter"])

print(f"Rows after cleaning: {len(df)}")
```

```python
# Filter to Oslo county only
df_oslo = df[df["Fylke"] == "Oslo"]

print(f"Oslo rows: {len(df_oslo)}")
print("Oslo schools:", df_oslo["Skole"].unique())
```

```python
# Separate standpunkt and written exam grades
standpunkt = df_oslo[df_oslo["Vurderingsform"] == "Standpunkt"]
skriftlig = df_oslo[df_oslo["Vurderingsform"] == "Skriftlig eksamen"]
muntlig = df_oslo[df_oslo["Vurderingsform"] == "Muntlig eksamen"]

# Calculate average grade per assessment type across all Oslo schools and subjects
avg_standpunkt = standpunkt["Snittkarakter"].mean()
avg_skriftlig = skriftlig["Snittkarakter"].mean()
avg_muntlig = muntlig["Snittkarakter"].mean()

print(f"Oslo average standpunkt grade:      {avg_standpunkt:.2f}")
print(f"Oslo average written exam grade:    {avg_skriftlig:.2f}")
print(f"Oslo average oral exam grade:       {avg_muntlig:.2f}")
print(f"\nGap standpunkt - written exam:      {avg_standpunkt - avg_skriftlig:.2f}")
print(f"Gap oral exam - standpunkt:         {avg_muntlig - avg_standpunkt:.2f}")
```

```python
# Calculate average standpunkt vs written exam gap per subject
sp_by_subject = standpunkt.groupby("Fag")["Snittkarakter"].mean()
sk_by_subject = skriftlig.groupby("Fag")["Snittkarakter"].mean()

# Merge the two series so we can compute the gap
gap_by_subject = (sp_by_subject - sk_by_subject).dropna().sort_values(ascending=False)

print("Standpunkt minus written exam gap per subject (Oslo):")
print(gap_by_subject)
```

```python
# Bar chart: grade gap per subject in Oslo
plt.bar(gap_by_subject.index, gap_by_subject.values)

plt.title("Standpunkt minus Written Exam Gap per Subject (Oslo)")
plt.xlabel("Subject")
plt.ylabel("Grade difference")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

```python
# Calculate average standpunkt vs written exam gap per school
sp_by_school = standpunkt.groupby("Skole")["Snittkarakter"].mean()
sk_by_school = skriftlig.groupby("Skole")["Snittkarakter"].mean()

gap_by_school = (sp_by_school - sk_by_school).dropna().sort_values(ascending=False)

print("Standpunkt minus written exam gap per school (Oslo):")
print(gap_by_school)
```

```python
# Bar chart: grade gap per school in Oslo
# This shows which schools are most lenient (high gap) or strict (low gap)
plt.bar(gap_by_school.index, gap_by_school.values)

plt.title("Standpunkt minus Written Exam Gap per School (Oslo)")
plt.xlabel("School")
plt.ylabel("Grade difference")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

```python
# Compare national average vs Oslo average for standpunkt and written exam
df_national = df[df["Fylke"] == "Nasjonalt"]

nat_sp = df_national[df_national["Vurderingsform"] == "Standpunkt"]["Snittkarakter"].mean()
nat_sk = df_national[df_national["Vurderingsform"] == "Skriftlig eksamen"]["Snittkarakter"].mean()

print(f"National average standpunkt:     {nat_sp:.2f}")
print(f"National average written exam:   {nat_sk:.2f}")
print(f"National gap:                    {nat_sp - nat_sk:.2f}")
print()
print(f"Oslo average standpunkt:         {avg_standpunkt:.2f}")
print(f"Oslo average written exam:       {avg_skriftlig:.2f}")
print(f"Oslo gap:                        {avg_standpunkt - avg_skriftlig:.2f}")
```

```python
# Bar chart: national vs Oslo standpunkt and written exam averages
labels = ["Standpunkt", "Written exam"]
national_vals = [nat_sp, nat_sk]
oslo_vals = [avg_standpunkt, avg_skriftlig]

x = range(len(labels))

plt.bar([i - 0.2 for i in x], national_vals, width=0.4, label="National")
plt.bar([i + 0.2 for i in x], oslo_vals, width=0.4, label="Oslo")

plt.xticks(x, labels)
plt.title("National vs Oslo: Average Grades by Assessment Type")
plt.xlabel("Assessment type")
plt.ylabel("Average grade")
plt.legend()

plt.tight_layout()
plt.show()
```

## Findings

- Standpunkt grades are consistently higher than written exam grades, both nationally and in Oslo.
- Oral exam grades are higher than standpunkt grades.
- The gap varies by subject – mathematics shows the largest standpunkt-exam gap.
- Oslo schools follow the same pattern as described in the Aftenposten article, but there is variation between individual schools.
